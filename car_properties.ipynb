```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mpg.csv')
df.head(2)
```

### What is the relationship between the power of a car's engine and its fuel efficiency? And how does this relationship vary by the number of cylinders the car has?

```python
sns.relplot(x="horsepower", y="mpg", data=df, kind="scatter", size="cylinders", hue='cylinders')
```

Cars with higher horsepower tend to get a lower number of miles per gallon. They also tend to have a higher number of cylinders.

### Does properties - mpg and acceleration vary by country of origin?

```python
sns.relplot(x='acceleration', y='mpg', data=df, hue='origin', style='origin', kind='scatter')
```

Cars from the USA tend to accelerate more quickly and get lower miles per gallon compared to cars from Europe and Japan.

### How has the average miles per gallon achieved by these cars changed over time? 

```python
sns.relplot(x='model_year', y = 'mpg', data=df, kind='line')
```

### Visualize distribution of miles per gallon for all the cars in each year

```python
sns.relplot(x="model_year", y="mpg", data=df, kind="line", ci='sd')
```

### How has the average horsepower for cars changed over time? And does this trend differ by country of origin?

```python
sns.relplot(x='model_year', y='horsepower', data=df, ci=None, kind='line', style='origin', hue='origin', markers = True, dashes=False)
```

Downward trend in horsepower was more pronounced among cars from the USA.
